# Conditional Statements

```python
# Checks if (true == TRUE)

a = -3
b = 5
c = 0

if a:
  print('a is valid')
else:
  print('a is not valid')

if b:
  print('b is valid')
else:
  print('b is not valid')

if c:
  print('c is valid')
else:
  print('c is not valid')
```

```python
# Checks if (d == True(1) )
# Explicitly checking

d = -3
e = 5
f = 1

if d == True:
  print('d is valid')
else:
  print('d is not valid')

if e == True:
  print('e is valid')
else:
  print('e is not valid')

if f == True:
  print('f is valid')
else:
  print('f is not valid')
```

# Membership Operator

```python
# <:> is the membership operator
# <in> & <not in>

stu = ['sam', 'tom', 'jerry' , 'steve' , 'pam', 'steve' , 'bruce', 'tony']

print('steve' in stu)
print('jerry' in stu)
print('steve' not in stu)
print('tony' not in stu)

# Length of array
print(len(stu))

# Sort in alphabetical, order then show max or min
print(max(stu))
print(min(stu))

# Alphabetical sorting
print(sorted(stu))
```

# Dictionary

- get() : Looks value in dictionary and returns 'NULL' if not present or can be configured.

```python
elements = {'hydrogen': {'number': 1, 'weight': 1.00794, 'symbol': 'H'},
            'helium': {'number': 2, 'weight': 4.002602, 'symbol': 'He'},
            'oxygen': {'number': 8, 'weight': 15.999, 'symbol': 'O'}}

# before
print(elements)

# adding entity & assigning value into dictionary
elements['hydrogen']['is_noble_gas'] = False
elements['helium']['is_noble_gas'] = True

# after
print(elements)
```

```python
print(elements['hydrogen']['is_noble_gas'])
```

```python
print(elements['helium']['is_noble_gas'])
```

# Loops

### Range Loop

```python
# range(start, stop, step)

for i in range(1, 10, 2):
  print(i)
```

### For Loop

```python
# for <variable>/<condition> in <range(<parameter>)>:
#   <operation>

# used for modification
for i in stu:
  print(i)

print('\n')

# used for viewing
for j in range(len(stu)):
  print(stu[j])
```

# While Loop

```python
# Pass statement : used to add delay

print('Hello')

for i in range(1000000000):
  pass

print('World')
```
