## Largest Number

````python
def max_number(num1, num2, num3):
  maximum = max(num1, num2, num3)
  return maximum

# Input
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))

# Calling function7
result = max_number(num1, num2, num3)

# Result
print("The maximum number is:", result)
````

## Guessing game

````python
import random

def guess_the_number():
  number = random.randint(1, 10)
  guess = 0
  attempts = 0

  while guess != number:
    guess = int(input("Guess a number between 1 and 10: "))
    attempts += 1

    if guess < number:
      print("Too low!")
    elif guess > number:
      print("Too high!")

  print("Congratulations! You guessed the number in",attempts,"attempts.")

guess_the_number()
````

## Check if array is sorted

````python
def is_sorted(arr):
  for i in range(1, len(arr)):
    if arr[i-1] > arr[i]:
      return False
  return True

sorted_array = [1, 2, 3, 4, 5]
unsorted_array = [1, 2, 1, 4, 5]

print(is_sorted(sorted_array))
print(is_sorted(unsorted_array))
````


## List Comprehension

>Syntax:
```
newlist = [<expression> for <element> in <iterable> if <condition == True>]
newlist = [x for x in fruits if x != "apple"]
```
>Parameter:
- <u>Expression</u>: Represents the operation you want to execute on every item within the iterable.
- <u>Element</u>: The term “variable” refers to each value taken from the iterable.
- <u>Iterable</u>: specify the sequence of elements you want to iterate through.(e.g., a list, tuple, or string).
- <u>Condition</u>: (Optional) A filter helps decide whether or not an element should be added to the new list.

>Return:The return value of a list comprehension is a new list containing the modified elements that satisfy the given criteria.

>Python List comprehension provides a much more short syntax for creating a new list based on the values of an existing list.

> Iterable : incrementable or decrementable.

> We can apply loop in iterable item.

````python
numbers = [3,4,5,6,7]

squared_numbers = [num**2 for num in numbers]

print(squared_numbers)
````

````python
languages = ['java', 'python', 'php', 'c', 'javascript']

languages_starting_with_p = [language for language in languages if language.startswith('p')]

print(languages_starting_with_p)
````

# Lambda Function

> A ***lambda function*** is a small <mark>anonymous function</mark> .

> A ***lambda function*** can take <mark>any number of arguments</mark> , but can <mark>only have one expression</mark> .

> Used for simple small functions needed for a short period of time.

> Used when functions are needed to be <mark>passed as agrument(input)</mark> to some other function.


Syntax :
```
lambda <arguments> : <expression>
or
lambda <input> : <output>
```

> argument is also called parameter.

> lambda functions don't have a name.

> Lambda fucntions can be <mark>called using the variable that  stores it</mark> .

````python
#EXAMPLE 1

x = lambda a : a + 10
print(x(5))
````

````python
# Example 2

x = lambda a, b : a * b
print(x(5, 6))
````

# Spiral Matrix (Homework)

Q ) Given an m x n matrix, return all elements of the matrix in spiral order.



Example 1:

image.png

Input: matrix = [[1,2,3],[4,5,6],[7,8,9]]

Output: [1,2,3,6,9,8,7,4,5]

````python
def spiralOrder(matrix):
    if not matrix:
        return []

    result = []
    top, bottom = 0, len(matrix) - 1
    left, right = 0, len(matrix[0]) - 1

    while top <= bottom and left <= right:
        # Traverse from left to right along the top row
        for i in range(left, right + 1):
            result.append(matrix[top][i])
        top += 1

        # Traverse from top to bottom along the right column
        for i in range(top, bottom + 1):
            result.append(matrix[i][right])
        right -= 1

        if top <= bottom:
            # Traverse from right to left along the bottom row
            for i in range(right, left - 1, -1):
                result.append(matrix[bottom][i])
            bottom -= 1

        if left <= right:
            # Traverse from bottom to top along the left column
            for i in range(bottom, top - 1, -1):
                result.append(matrix[i][left])
            left += 1

    return result

# Example usage
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(spiralOrder(matrix))
````
