# K-Means Clustering

- K = no. of clusters
- 'K' is not known & it is decided by the algorithm.


2 types of methods used to decide optimim K value
1) Elbow Method
2) Purpose Based Method

```python
# Importing Libraries
import numpy as numpy
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# Importing Dataset
data = pd.read_csv('Mall_Customers.csv')
data
```

```python
data.info()
```

```python
data.describe(include='all')
```

```python
# Creating a scatter plot

plt.scatter(data['Annual Income (k$)'],data['Spending Score (1-100)'])
plt.title("Scatter Plot")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

```python
X = data.iloc[:,[3,4]]
X.head()
```

```python
# Importing KMeans class from Scikit- Learn
from sklearn.cluster import KMeans
```

```python
# Defining the number of clusters
kmeans = KMeans(2, random_state = 10)
```

```python
kmeans.fit(X)
```

```python
# Checking the predicted clusters
kmeans.labels_
```

```python
# Checking the cluster centroids

kmeans.cluster_centers_
```

```python
data_new = data.copy()
data_new['cluster'] = kmeans.labels_
data_new
```

```python
plt.scatter(data_new['Annual Income (k$)'], data_new['Spending Score (1-100)'], c=data_new['cluster'], cmap='rainbow', alpha=0.5)

plt.plot(kmeans.cluster_centers_[0,0], kmeans.cluster_centers_[0,1], c='k', marker='o')
plt.plot(kmeans.cluster_centers_[1,0], kmeans.cluster_centers_[1,1], c='k', marker='o')


plt.title('Scatter plot')
plt.xlabel('Annual Income(in k$)')
plt.ylabel('Spending Score')
plt.show()
```

```python
# plotting WCSS for different no. of clusters

wcss = []
for i in range(1,20):
    kmeans = KMeans(i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,20), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans.cluster_centers_
```

```python
data_new = data.copy()
data_new['Clusters'] = kmeans.labels_
data_new.head(10)
```

```python
plt.scatter(data_new['Annual Income (k$)'], data_new['Spending Score (1-100)'], c=data_new['Clusters'], cmap='rainbow', alpha=0.5)

plt.plot(kmeans.cluster_centers_[0,0],kmeans.cluster_centers_[0,1],marker='o',c='k')
plt.plot(kmeans.cluster_centers_[1,0],kmeans.cluster_centers_[1,1],marker='o',c='k')
plt.plot(kmeans.cluster_centers_[2,0],kmeans.cluster_centers_[2,1],marker='o',c='k')
plt.plot(kmeans.cluster_centers_[3,0],kmeans.cluster_centers_[3,1],marker='o',c='k')
plt.plot(kmeans.cluster_centers_[4,0],kmeans.cluster_centers_[4,1],marker='o',c='k')

plt.xlabel('Annual Income ')
plt.ylabel('Spending Score')
plt.title('Scatter Plot')
plt.xlabel('Annual Income ')
plt.ylabel('Spending Score')
plt.show()
```

```python
data_new.groupby('Clusters').mean(numeric_only = True)
```

```python
X = data.loc[:,['Annual Income (k$)','Spending Score (1-100)']]
X.head()
```

```python
wcss = []
for i in range (1,11):
    kmeans = KMeans(i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(5, random_state=12)
kmeans.fit(X)
```

```python
centroids = kmeans.cluster_centers_
centroids
```

```python
kmeans.labels_
```

```python
data_new = data.copy()
data_new['Clusters'] = kmeans.labels_
data_new.head(10)
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure(figsize=(10,6))
ax = plt.axes(projection='3d')
ax.view_init(20,10)

ax.scatter3D(data_new['Age'], data_new['Annual Income (k$)'], data_new['Spending Score (1-100)'], c=data_new['Clusters'], cmap='rainbow', alpha = 0.7)

ax.plot3D(centroids[0,0], centroids[0,1], centroids[0,2], c='k', marker='o')
ax.plot3D(centroids[1,0], centroids[1,1], centroids[1,2], c='k', marker='o')
ax.plot3D(centroids[2,0], centroids[2,1], centroids[2,2], c='k', marker='o')
ax.plot3D(centroids[3,0], centroids[3,1], centroids[3,2], c='k', marker='o')
ax.plot3D(centroids[4,0], centroids[4,1], centroids[4,2], c='k', marker='o')
plt.show()
```
