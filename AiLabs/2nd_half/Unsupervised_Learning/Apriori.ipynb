## Association Rule Learning
Apriori Algorithm is  used to find association rules.

```python
!pip install apyori 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Transactions_1=[['eggs','bacon','soup'],
                ['eggs','bacon','apple'],
                ['soup','bacon','banana']]
Transactions_1
```

```python
print('support of {egg,soup}:',1/3,'onf of {egg,soup}:',1/2,'lift of {egg,soup}:',(1/(2))/(2/3) )
```

```python
from apyori import apriori
rules = apriori(Transactions_1, min_support = 1, min_confidence = 0.5, min_lift = 0.7)
#making rules for two items
#Visualizing the results

results = list(rules)
results
```

```python
dataset = pd.read_csv("Market_Basket_Optimisation.csv", header = None)
ds = dataset.shape
ds
```

```python
dataset
```

```python
transaction = []
for i in range(0,ds[0]):
  transaction.append([str(dataset.values[i,j]) for j in range(0,20)])

transaction
```

```python
rules = apriori(transaction, min_support = 0.002, min_confidence = 0.001, min_lift = 0.2)
results = list(rules)
results
```
